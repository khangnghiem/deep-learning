```python
import torchvision
from tqdm import tqdm
from src.config.paths import BRONZE, GOLD

EXPERIMENT = "005_svhn_cnn"

def extract_to_gold(split_name, tv_split):
    out_dir = GOLD / EXPERIMENT / split_name
    dataset = torchvision.datasets.SVHN(root=str(BRONZE / "svhn"), split=tv_split, download=True)
    for i in range(10):
        (out_dir / str(i)).mkdir(parents=True, exist_ok=True)
    print(f"Writing {len(dataset)} {split_name} images...")
    for i in tqdm(range(len(dataset))):
        img, label = dataset[i]
        img.save(str(out_dir / str(label) / f"{i:05d}.png"))

extract_to_gold("train", "train")
extract_to_gold("val", "test")
print("✅ Gold data creation complete!")
```
