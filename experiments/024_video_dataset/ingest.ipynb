```python
# Cell 3: Data logic
import os
import subprocess
import shutil

landing_dir = '/content/drive/MyDrive/data_lake/00_landing/polyp/ldpolypvideo'
silver_dir = '/content/drive/MyDrive/data_lake/01_silver/polyp/ldpolypvideo'
os.makedirs(silver_dir, exist_ok=True)

print(f"Extracting data to {silver_dir}")
train_valid_path = os.path.join(landing_dir, 'TrainValid.rar')
test_path = os.path.join(landing_dir, 'Test.rar')
videos_zip_path = os.path.join(landing_dir, 'videos with polyps.zip')

# Using unrar for rar files
if os.path.exists(train_valid_path):
    if not os.path.exists(os.path.join(silver_dir, 'TrainValid')):
        print('Extracting TrainValid.rar...')
        subprocess.run(['unrar', 'x', '-o+', train_valid_path, silver_dir], check=True)
    else:
        print('TrainValid already extracted.')

if os.path.exists(test_path):
    if not os.path.exists(os.path.join(silver_dir, 'Test')):
        print('Extracting Test.rar...')
        subprocess.run(['unrar', 'x', '-o+', test_path, silver_dir], check=True)
    else:
        print('Test already extracted.')

# Using unzip for zip files
if os.path.exists(videos_zip_path):
    if not os.path.exists(os.path.join(silver_dir, 'videos with polyps')):
        print('Extracting videos with polyps.zip...')
        subprocess.run(['unzip', '-q', '-o', videos_zip_path, '-d', silver_dir], check=True)
    else:
        print('Videos already extracted.')

print('Extraction complete!')
```
