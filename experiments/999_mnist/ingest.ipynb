```python
import torchvision
from tqdm import tqdm
from src.config.paths import BRONZE, GOLD

EXPERIMENT = "999_mnist"

def extract_to_gold(train=True):
    split = "train" if train else "val"
    out_dir = GOLD / EXPERIMENT / split
    dataset = torchvision.datasets.MNIST(root=str(BRONZE / "mnist"), train=train, download=True)
    classes = dataset.classes
    for cls in classes:
        (out_dir / cls).mkdir(parents=True, exist_ok=True)
    print(f"Writing {len(dataset)} {split} images to {out_dir}...")
    for i in tqdm(range(len(dataset))):
        img, label_idx = dataset[i]
        label = classes[label_idx]
        img.save(str(out_dir / label / f"{i:05d}.png"))

extract_to_gold(train=True)
extract_to_gold(train=False)
print("✅ Gold data creation complete!")
```
