This notebook runs on **CPU**. It extracts data into standard ImageFolder in the Gold layer.


```python
# Cell 1: Colab Environment Setup
from google.colab import drive, runtime
drive.mount("/content/drive")

# Clone code repos from GitHub
!git clone --depth 1 -q https://github.com/khangnghiem/deep-learning.git /content/deep-learning

import os, sys
sys.path.insert(0, "/content/deep-learning")
os.chdir("/content/deep-learning/experiments/008_blood_cells_cnn")
```

```python
# Bronze data is already in ImageFolder format — copy to Gold
import shutil
from src.config.paths import BRONZE, GOLD

bronze_dir = BRONZE / "blood_cells"
gold_dir = GOLD / "008_blood_cells_cnn"

for split in ["train", "val"]:
    src = bronze_dir / split
    dst = gold_dir / split
    if src.exists() and not dst.exists():
        shutil.copytree(str(src), str(dst))
        print(f"Copied {src} -> {dst}")
    elif dst.exists():
        print(f"{dst} already exists, skipping.")
    else:
        print(f"WARNING: {src} not found!")
print("✅ Gold data creation complete!")
```
