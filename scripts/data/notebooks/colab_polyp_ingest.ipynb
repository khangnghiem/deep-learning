```python
# Cell 3: Download all Fast-Diag polyp datasets

POLYP_DATASETS = [
    'kvasir-seg',
    'cvc-clinicdb',
    'cvc-colondb',
    'bkai-igh',
    'etis-larib',
    'polypdb',
    'ldpolypvideo',
]

results = {}
for ds in POLYP_DATASETS:
    print(f"\n{'='*60}")
    print(f"  Downloading: {ds}")
    print(f"{'='*60}")
    try:
        result = download_dataset(ds)
        results[ds] = 'success' if result else 'failed'
    except Exception as e:
        results[ds] = f'error: {str(e)[:100]}'
        print(f'ERROR: {e}')

print(f"\n\n{'='*60}")
print('SUMMARY')
print(f"{'='*60}")
for ds, status in results.items():
    icon = '✅' if status == 'success' else ('⏭️' if status == 'skipped' else '❌')
    print(f'{icon} {ds}: {status}')
```

```python
# Cell 4: Verify downloads
from pathlib import Path

DRIVE = Path('/content/drive/MyDrive/data_lake')
expected = [
    '01_bronze_medical/kvasir-seg',
    '01_bronze_medical/cvc-clinicdb',
    '01_bronze_medical/cvc-colondb',
    '01_bronze_medical/bkai-igh',
    '01_bronze_medical/etis-larib',
    '01_bronze_medical/polypdb',
    '01_bronze_medical/ldpolypvideo',
]

print('\nVerification:')
for ds in expected:
    path = DRIVE / ds
    if path.exists():
        files = list(path.rglob('*'))
        total_size = sum(f.stat().st_size for f in files if f.is_file())
        print(f'✅ {ds} — {len(files)} files, {total_size/1024/1024:.1f} MB')
    else:
        print(f'❌ {ds} — NOT FOUND')
```
